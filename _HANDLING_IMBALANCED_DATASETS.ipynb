#HANDLING IMBALANCED DATASETS
A dataset with an unqual class distrbution

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
credit_card_data= pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#distribution of two classes
credit_card_data['Class'].value_counts()
```

```python
# this is highly imbalanced dataset
#0-legit transactions
#1- fraudalent transactions
```

```python
#seperating the legit and fraudlentdat
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

#Under-Sampling
it is a technique to handle imbalace data

```python
# taking 153 legit transactions out of 53417 datas same(similar) as that of fraudalent transaction...i,e fraud
#taking the random sampling
```

```python
legit_sample=legit.sample(n=153)
```

```python
legit_sample.shape
```

```python
#concatenate the 2 dataframes
#1. fraud
#2. legit_sample
new_dataset=pd.concat([legit_sample,fraud],axis=0) # axis=0 addition row wise for columnwise axis=1
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```
